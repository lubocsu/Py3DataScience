```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. axes_style and set_style

```python
%matplotlib inline
```

```python
x = np.linspace(0,14,100)
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x, y1)
    plt.plot(x, y2)
```

```python
sinplot()
```

```python
import seaborn as sns
```

```python
sinplot()
```

```python
style = ['darkgrid', 'dark', 'white','whitegrid', 'ticks']
```

#### 定义好的五种主题，set_style方法设置主题

```python
sns.set_style(style[3], {'grid.color': 'red'})
```

```python
sinplot()
```

```python
sns.axes_style()
```

这些参数都是可以微调的。

```python
sinplot()
```

```python
sns.set()
```

重置为原来的风格

```python
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
sns.set()
```

```python
sinplot()
```

### 2. plotting_context() and set_context() 

>更改里面曲线的属性，粗细之类

```python
context = ['paper', 'notebook', 'talk', 'poster']
```

```python
sns.set_context(context[0], rc={'grid.linewidth': 3.0})
```

```python
sinplot()
```

```python
sns.set()
```

```python
sns.plotting_context()
```

plotting_context 这些context中的信息都可以进行配置
