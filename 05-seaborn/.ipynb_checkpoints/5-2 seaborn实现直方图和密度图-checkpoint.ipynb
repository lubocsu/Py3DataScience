```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
%matplotlib inline
```

```python
import seaborn as sns
```

```python
s1 = Series(np.random.randn(1000))
```

```python
plt.hist(s1,rwidth=0.9)
```

```python
s1.plot(kind='kde')
```

distplot a是要画的数据，bins是分多少个区间，hist是是否去画直方图。

kde=true会同时画出来密度图。

```python
sns.distplot(s1, bins=200, hist=True, kde=False, rug=True)
```

rug设置为true，会画出一些小的竖线，代表具体的分布

```python
sns.kdeplot(s1, shade=True, color='g')
```

kdeplot 画密度图 shade是对曲线是否进行填充

在seaborn中直接调用Matplotlib的plt方法。
