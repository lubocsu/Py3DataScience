```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import Series, DataFrame
```

sns中的内置数据，会在线去下载数据。

```python
df = sns.load_dataset('flights')
```

```python
df = df.pivot(index='month', columns='year', values='passengers')
```

#### 热力图

```python
sns.heatmap(df)
```

```python
df.plot()
```

- annot参数: 显示实际的数
- fmt参数: 让它格式化为整数

```python
sns.heatmap(df, annot=True, fmt='d')
```
