```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
df = DataFrame(
    np.random.randint(1,10,40).reshape(10,4),
    columns=['A','B','C','D']
)
df.head()
```

randint 1到10的数据取40个

```python
df.plot(kind='bar')
```

每列画一条线或一个图

```python
plt.show()
```

```python
df.plot(kind='barh')
plt.show()
```

barh 横向显示的柱状图

```python
df.plot(kind='bar', stacked=True)
```

对于同一个columns下的数据进行叠加。

```python
plt.show()
```

```python
df.plot(kind='area')
```

```python
plt.show()
```

#### 对于所有行进行一起画图

```python
for i in df.index:
    # 取第i行画线，label为i
    df.iloc[i].plot(label=str(i))
plt.legend()
plt.show()
```
