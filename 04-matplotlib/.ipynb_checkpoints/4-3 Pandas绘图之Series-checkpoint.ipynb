```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
s =Series([1,2,3,4,5])
s
```

```python
s.cumsum()
```

cumsum()是一种累加的求和。

```python
s1 = Series(np.random.randn(1000)).cumsum()
s1.head()
```

```python
s2 = Series(np.random.randn(1000)).cumsum()
s2.head()
```

```python
s1.plot()
```

```python
plt.show()
```

- kind 默认值是line，也就是画一个线性图。
- grid参数会显示一些方格

```python
s1.plot(kind='line',grid=True, label='S1', title='This is Series')
s2.plot(label='S2')
```

```python
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
```

```python
ax
```

```python
ax[0].plot(s1)
```

```python
ax[1].plot(s2)
```

```python
plt.show()
```

#### 还可以通过series的plot方法调用

```python
fig, ax = plt.subplots(2,1)
```

```python
s2.plot(ax=ax[0], label='S2')
s1[0:10].plot(ax=ax[1], label='S1',kind='bar')
```

```python
plt.legend()
plt.show()
```
