在一个图中画出多组数据的图

```python
t = np.arange(0.0, 2.0, 0.1)
t
```

从0到2，以间隔0.1生成数据

```python
t.size
```

```python
s = np.sin(t*np.pi)
s
```

```python
s.size
```

```python
plt.plot(t,s,'r--',label='aaaa')
plt.plot(t*2, s, 'b--', label='bbbb')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.title('this is a demo')
plt.legend()
```

xlabel ylabel 设置x轴y轴的label。title和线的图注。

plt.legend() 显示样例。
