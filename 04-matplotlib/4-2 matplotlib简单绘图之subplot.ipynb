```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0, 5.0)
print(x)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

通过linspace 生成等差间隔的数据

```python
plt.plot(x, y1, 'b--', label='sin(pi*x)')
plt.ylabel('y1 value')
plt.plot(x, y2, 'r--', label='sin(pi*2x)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('this is x-y value')
plt.legend()
```

```python
plt.show()
```

### 画子图

```python
plt.subplot(2,1,1)
plt.plot(x, y1, 'b--')
plt.ylabel('y1')
plt.subplot(2,1,2)
plt.plot(x, y2, 'r--')
plt.ylabel('y2')
plt.xlabel('x')
plt.show()
```

```python
# plt.subplot(2,1,1) 生成两行一列的图。第三个数指位置
plt.subplot(221)
plt.plot(x, y1, 'b--')
plt.ylabel('y1')
plt.subplot(222)
plt.plot(x, y2, 'r--')
plt.ylabel('y2')
plt.xlabel('x')
plt.subplot(223)
plt.plot(x, y1, 'r*')
plt.subplot(224)
plt.plot(x, y1, 'b*')
```

```python
plt.show()
```

#### 另外一种画子图的方式

```python
figure, ax = plt.subplots(2,2)
ax[0][0].plot(x, y1)
ax[0][1].plot(x, y2)
plt.show()
```
