### 查看中间的某几行到某几行的数据。

`sub_df.iloc[10:20,:]`

>对于行取10到19行，列并不做过滤:也就是显示所有列。

````python
sub_df.iloc[10:20,:]
````

iloc函数除过过滤行:

- 当然也可以过滤列。也是通过列的index来过滤的。我们看到的名字其实是列的label，而不是列的index。

```
sub_df.iloc[10:20,0:2]
```

>对于行取10到19行，列并取第0到1列。

````python
sub_df.iloc[10:20,0:2]
````

它是通过index来实现的，所以与具体的label名是没有关系的。

````python
tmp_df = sub_df.iloc[10:20,0:2]
````

````python
tmp_df
````

````python
tmp_df.iloc[2:4,:]
````

>可以看到: 这里的index并不是label的字面值，而是index值。

有时我们想要通过index去过滤，有时候我们是需要通过name(label值)去过滤的。

- 列数特别多， 我们想要通过我们的label值来过滤

>`iloc`不加`i`(index)的时候是通过label的name来定位的。

````python
tmp_df.loc[15:17,:]
````

>上面代码中的15到17不是index了，而是我们的label name

````python
tmp_df.loc[15:17,:'director_name']
````

>从任意label直到director_name的列返回回来

### 总结

- iloc 是基于index的，这个index是隐藏的。
- loc 是基于我们可以看见的label的，这个label就是我们可以看到的。
