### 如何重命名dataframe的index


````python
df1 = DataFrame(np.arange(9).reshape(3,3), index=['BJ','SH','GZ'],columns=['a','b','c'])
df1
````

````python
df1.index = ['A','B','C']
df1
````

````python
df1.index = Series(['bj','sh','gz'])
df1
````

````python
df1.index
````

````python
df1.index = df1.index.map(str.upper)
df1
````

Apply mapper function to an index.

运行这个函数对于每一个index

````python
df1.rename(index=str.lower,columns=str.upper)
````

#### 传入字典对于特定的columns，或index进行更改

````python
df1.rename(index={'BJ':'beijing'},columns={'a':'A','b':'B'})
````

#### 自定义一个自己的map函数

````python
df1
````

````python
def test_map(x):
    return x + '_mtianyan'
````
