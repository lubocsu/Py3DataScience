透视表的概念来自于excel

我们为了更好的展示我们的数据，临时把我们原始的表结构做了一些变更。原始的行和列做了一些变化。

根据这些变化，数据也做了一点变化。

````python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
````

一共有8个columns。销售信息: 账户，客户姓名，销售代表，老板，产品，数量，单价，状态

#### pandas生成透视表pivot_table

```
pd.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All')
```

- index是我们具体要对哪几个columns做透视。

````python
pd.pivot_table(df, index=['Name'])
````

生成了一个新的dataframe，这个dataframe的name这一列是unique，做了去重的。

对于其他列是使用了默认的agg_func 默认是mean

销售代表和Manager之前有一个一对多的关系。

- values指定要看那几列值
- 对于数据看不同的columns
- fill_value 将缺失值进行填充

````python
pd.pivot_table(df, index=['Manager','Rep'],values=['Price','Quantity'],columns=['Product'], fill_value=0, aggfunc='sum')
````

透视表可以实现一个强大的数据分析功能

- 前提: 我们要对数据有一定的了解。

>每个columns是什么意思，不同columns之间的关系
